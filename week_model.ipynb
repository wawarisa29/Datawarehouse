```
#SVC ---> à¸„à¸·à¸­ RBF
# Tune Model and Select Model based on Training(CV) Accuracy
from sklearn.svm import SVC 
# trainning accuracy /  à¹„à¸”à¹‰à¸„à¹ˆà¸²à¸•à¸±à¸§à¹€à¸¥à¸‚
# à¸§à¸´à¸˜à¸µà¸«à¸²à¸•à¸±à¸§ training accuracy 
from sklearn.model_selection import cross_val_score
clf = SVC(gamma='scale')
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()

```

```
#K-Nearest
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
clf = KNeighborsClassifier(n_neighbors=5)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
gnb = GaussianNB()
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# Decision Tree
from sklearn import tree
from sklearn.model_selection import cross_val_score
clf = tree.DecisionTreeClassifier(criterion='entropy')
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# Support Vector Machine
from sklearn import svm
from sklearn.model_selection import cross_val_score
clf = svm.SVC(gamma='scale')
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# SGD Classifier
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
clf = linear_model.SGDClassifier(max_iter=1000, tol=1e-3)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
#Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
clf = RandomForestClassifier(n_estimators=200, max_depth=9,random_state=0)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# XGBoots
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
clf = XGBClassifier(n_estimate=50,max_depth=8)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

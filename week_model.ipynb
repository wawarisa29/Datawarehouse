```
from sklearn.model_selection import train_test_split 
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.25, random_state=0)
#test = val มันเป็นเทรนเฉยๆๆ
```

```
#ต่อไปคือ Normalization
from sklearn.preprocessing import MinMaxScaler # Model sklearn
scaler = MinMaxScaler(feature_range=(0,1)) # ประเภท Normalization
scaler.fit(x_train) #สร้างโมเดลการ โนมอไลซ์ จาก Trainning data
x_train_norm = scaler.transform(x_train) #สั่งให้ โมเดล แปลงสเกลที่ต้องการ
x_val_norm = scaler.transform(x_val)
```

```
x_train_norm.max(axis=0) #เช็คmax
```

```
# สร้าง Model
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(C=15, intercept_scaling= 0.5)
clf
#แนะนำให้ set ค่า C ---> search ได้
```

```
#สั่ง train model
clf.fit(x_train_norm,y_train)
```

```
# predict data
# ทำนายเเล้วเก็บไว้ที่ตัวแปร
y_predict = clf.predict(x_val_norm)
y_predict
```

```
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
clf = LogisticRegression(C=15, intercept_scaling= 0.5)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
#SVC ---> คือ RBF
# Tune Model and Select Model based on Training(CV) Accuracy
from sklearn.svm import SVC 
# trainning accuracy /  ได้ค่าตัวเลข
# วิธีหาตัว training accuracy 
from sklearn.model_selection import cross_val_score
clf = SVC(gamma='scale')
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()

```

```
#K-Nearest
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
clf = KNeighborsClassifier(n_neighbors=5)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
gnb = GaussianNB()
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# Decision Tree
from sklearn import tree
from sklearn.model_selection import cross_val_score
clf = tree.DecisionTreeClassifier(criterion='entropy')
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# Support Vector Machine
from sklearn import svm
from sklearn.model_selection import cross_val_score
clf = svm.SVC(gamma='scale')
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# SGD Classifier
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
clf = linear_model.SGDClassifier(max_iter=1000, tol=1e-3)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
#Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
clf = RandomForestClassifier(n_estimators=200, max_depth=9,random_state=0)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```

```
# XGBoots
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
clf = XGBClassifier(n_estimate=50,max_depth=8)
accuracy = cross_val_score(clf,x_train_norm,y_train,cv=5)
accuracy.mean()
```
